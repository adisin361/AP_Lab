```python
df1 = pd.read_csv('C:\\Users\\KIIT\\Downloads\\housing.csv')
df1.describe()

scaler = MinMaxScaler()
df_norm = pd.DataFrame(scaler.fit_transform(df1), columns=df1.columns)
print("Normal Data")
df_norm
```

```python
std_scaler = StandardScaler()
std_scaler
df_std = pd.DataFrame(std_scaler.fit_transform(df1), columns=df1.columns)
print("Standard Data")
df_std
```

```python
corr=df1.corr()
ax = sns.heatmap(
    corr,
    vmin=-1, vmax=1, center=0,
    cmap=sns.diverging_palette(222,270,n=7),
    square=True,
    linewidths=.25
)

ax.set_xticklabels(
    ax.get_xticklabels(),
    ha='center',
    va='center',
);
```
